# Objetivo
* Gerar arquivo categorias
* Gerar sumario de pesquisas de cada categoria

```
# from google.colab import drive
# drive.mount('/content/drive')

import requests
import pandas as pd
import os.path

Url = 'https://servicodados.ibge.gov.br/api/v3/agregados?nivel=N6'
Data = requests.get(Url).json()
```

## Gerar arquivo categorias
* categorias.xlsx

```

nome = []
id =[] 
for i in Data:
  nome.append(i.get('nome',''))
  id.append(i.get('id',''))
  dt_categorias = pd.DataFrame([id,nome])
dt_categorias = dt_categorias.T
dt_categorias.columns = ['id','categoria']
dt_categorias.to_excel("./categorias.xlsx")
dt_categorias.head()
```

## Gerar sumario de pesquisas de cada categoria

```
"""
Passar [id]
Ler todas as pesquisas respectivas a id
Exportar em xlsx
"""



dados_id=[]
def id_pesquisas(id,Data):
    for i in Data:
        if i['id'] == id:
            dados_id = i['agregados']

    return dados_id

def export_pesquisas(lista_id,Data):
    for i in lista_id:
        dados = id_pesquisas(i,Data)
        arq = pd.DataFrame(dados)
        path = "./database/sumario/sumario_"+str(i)+".xlsx"
        arq.to_excel(path)
###
export_pesquisas(dt_categorias['id'],Data)
```
