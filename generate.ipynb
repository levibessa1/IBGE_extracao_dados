```python
import json
import time
#time.sleep(5)

def extrair_pesquisa(diretorio,id_pesquisa,id_cidade):#diretorio):
    arquivo = str(diretorio)+'/'+str(id_pesquisa)+'.json'
    if os.path.isfile(arquivo) :
        pass
    else:
        try:
            url = 'https://servicodados.ibge.gov.br/api/v3/agregados/'+str(id_pesquisa)+'/variaveis?localidades=N6['+str(id_cidade)+']'
            Data = requests.get(url).json()    
            with open(arquivo,'w') as f:
                json.dump(Data,f)
        except requests.exceptions.HTTPError as errh:
            print ("Http Error:",errh)
        except requests.exceptions.ConnectionError as errc:
            print("\n Arquivo>> {} excedeu o tempo de resposta".format(arquivo))
            print ("Error Connecting:",errc)
            arquivo = str(diretorio)+'/'+str(id_pesquisa)+'_erro'+'.txt'
            texto = str(errc)
            with open(arquivo,'w') as f:
                f.write(texto)
#        except requests.exceptions.Timeout as errt:
#            print ("Timeout Error:",errt)
#        except requests.exceptions.RequestException as err:
#            print ("OOps: Something Else",err)
            
def criar_dir_cidade(categoria,nome_cidade):
    diretorio = './database/'+str(categoria)+'/'+str(nome_cidade)
    os.makedirs(diretorio,exist_ok=True)
    return(diretorio)

def criar_dir_categoria(name):
    diretorio = './database/'+str(name)    
    os.makedirs(diretorio,exist_ok=True)

def export_sumarios(nameFile):#sumario_AA.xlsx
    nameFile = re.sub("_"," ",nameFile)
    nameFile = re.sub(".xlsx","",nameFile)
    categoria = nameFile.split()[1]
    print("***categoria =",categoria)
    data = pd.read_excel("./database/sumario/sumario_"+categoria+".xlsx").iloc[:,1:]
    
    criar_dir_categoria(categoria)
    
    for cidade in codigo_cidade: 
        print(">> Holding time to requests 1 min")
        time.sleep(60)
        print("\n {} a ser processada na categoria {}".format(cidade[0],categoria))          
        dir_cidade = criar_dir_cidade(categoria,cidade[0])
        for dt in data['id']:
            extrair_pesquisa(dir_cidade,id_pesquisa=dt,id_cidade=cidade[1]) 
```
